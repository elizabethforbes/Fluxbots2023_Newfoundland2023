### testing regression

```python
X = event.data['Unix.Epoch.Time']-event.data['Unix.Epoch.Time'].min()
y = (event.data['CO2_mass']-event.ambient)

df = pd.DataFrame(
{'X': X,
'y': y}
)

# write this dataframe to "tests" folder in master branch
df.to_csv("tests/NMWCOM2_event1079_forKelly.csv", index = False)
```

```python
ymean = np.mean(y)
xmean = np.mean(X)

# calculate terms needed for numerator and denominator of beta
df['xycov'] = (df['X'] - xmean) * (df['y'] - ymean)
df['xvar'] = (df['X'] - xmean)**2

# Calculate beta and alpha (below from paper draft)
# "the flux of CO$_2$ into the chamber for each measurement interval is determined 
# as the estimated initial rate of change of CO2 mass, B1 [kg/sec], divided by the 
# surface area of the chamber (m2)"
beta = df['xycov'].sum() / df['xvar'].sum()
alpha = ymean - (beta * xmean)
print(f'alpha = {alpha}')
print(f'beta = {beta}')
```

# linear regression, forcing regression through zero (aka no "b" in mx + b):

### then use to test Uncertainty in Slope Parameter in linear regression methods

```python
# define function lin_fit with no y-intercept:


def lin_fit(x, y):
    '''Fits a linear fit of the form mx+b to the data'''
    fitfunc = lambda params, x: params[0] * x                #create fitting function of form mx (no b)
    errfunc = lambda p, x, y: fitfunc(p, x) - y              #create error function for least squares fit

    init_a = 0.5                            #find initial value for a (gradient)
    #init_b = min(y)                         #find initial value for b (y axis intersection) (NO Y INTERCEPT HERE)
    #init_p = np.array((init_a, init_b))  #bundle initial values in initial parameters
    init_p = np.array((init_a))  #bundle initial values in initial parameters

    #calculate best fitting parameters (i.e. m) using the error function
    p1, success = scipy.optimize.leastsq(errfunc, init_p.copy(), args = (x, y))
    f = fitfunc(p1, x)          #create a fit with those parameters
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    n = np.size(x)
    ssxx = np.sum(x*x) - n*x_mean*x_mean # sum of squared deviations from the mean for X
    sse = np.sum((f - y)**2) # residual sum of squares of errors
    syx2 = sse/(n-2) # variance of y(x), predicted values of y at x versus actual values of y at x (using sse)
    sm2 = syx2 / ssxx
    CI_95 = 2*(math.sqrt(sm2))
    upperCI = p1 + CI_95
    lowerCI = p1 - CI_95

    return p1, upperCI, lowerCI # RETURNS: slope, and upper/lower 95% confidence intervals of slope
```

```python
# lin_fit(X, y)

linfit = lin_fit(X, y)

# array with slope, upperCI, and lowerCI
linfit
```
